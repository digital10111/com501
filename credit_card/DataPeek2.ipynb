```python
import pandas as pd
from matplotlib import pyplot
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("creditcard_dataset_2022.csv")
```

```python
df.Amount.describe()
```

```python
fig, axs = pyplot.subplots(1, 4, figsize=(14, 5))

sns.boxplot(x="Normal", y="Amount", data=df, ax=axs[0])

amount_l2500 = df.Amount < 2500
sns.boxplot(x="Normal", y="Amount", data=df[amount_l2500], ax=axs[1])

amount_l500 = df.Amount < 500
sns.boxplot(x="Normal", y="Amount", data=df[amount_l500], ax=axs[2])

amount_l250 = df.Amount < 180
sns.boxplot(x="Normal", y="Amount", data=df[amount_l250], ax=axs[3])
fig.tight_layout()
```
