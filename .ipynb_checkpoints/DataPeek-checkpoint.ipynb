```python
features = [col for col in df.columns if 'V' in col]
features.append("Amount")
features = np.asarray(features)
features = features.reshape(6, 5).tolist()
```

```python
fig, axs = pyplot.subplots(6, 5, figsize=(30, 20))

for i, feature_row in enumerate(features):
    for j, feature in enumerate(feature_row):
        sns.kdeplot(df[feature], hue=df.Normal, ax=axs[i][j])

fig.tight_layout()
```

```python
fig, ax = pyplot.subplots(figsize=(15, 10))
sns.heatmap(df.corr(), cmap="coolwarm", ax=ax)
```

```python
df.V1.fillna(df.V1.mean(), inplace=True)
cs = cosine_similarity(df.drop(["Amount", "Normal"], axis=1))
```
